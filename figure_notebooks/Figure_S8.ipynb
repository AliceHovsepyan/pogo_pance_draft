## Figure S8A: Uncorrected temporal nanopore data overtime of Figure 3C.

Data correspond to the temporal enrichment analysis presented in Figure 3C. Mutation frequencies are shown for each cycle by amino acid position. To reduce noise and correct for systematic sequencing artifacts, particularly those affecting certain positions such as R38, enrichment values were normalized by subtracting the mutation rates observed during the first mutagenesis cycle. This first cycle was assumed to reflect random mutagenesis plus persistent technical noise, enabling clearer resolution of biologically relevant mutation enrichments during subsequent selection steps. Mutation frequencies are displayed on a `0–0.25` scale, with residues exceeding `25%` enrichment shaded in <span style="color: yellow"><b>yellow</b></span>. 

The data underlying this visualization was generated from preprocessed reads.
